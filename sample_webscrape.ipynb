```
from bs4 import BeautifulSoup as bs
import requests
```

```
link ='https://www.tripadvisor.in/Hotel_Review-g304556-d1597314-Reviews-Hilton_Chennai-Chennai_Madras_Chennai_District_Tamil_Nadu.html#REVIEWS'
```

```
page = requests.get(link)
```

```
page
```

```
page.content
```

```
soup = bs(page.content,'html.parser')
```

```
print(soup.prettify())
```

```
names = soup.find_all('a',class_='ui_header_link _1r_My98y')
```

```
names
```

```
cust_names =[]
for i in range(0,len(names)):
  cust_names.append(names[i].get_text())
```

```
cust_names
```

```
Review_Title = soup.find_all('a', class_ ='ocfR3SKN')
```

```
Review_Title
```

```
Review_Titles=[]
for i in range(0,len(Review_Title)):
  Review_Titles.append(Review_Title[i].get_text())
```

```
Review_Titles
```

```
Review = soup.find_all('q',class_='IRsGHoPm')
```

```
Review
```

```
Reviews=[]
for i in range(0,len(Review)):
  Reviews.append(Review[i].get_text())
```

```
Reviews
```

```
import pandas as pd
df = pd.DataFrame()
```

```
df['Customer_Name'] = cust_names
df['Review_Titles'] =Review_Titles
df['Reviews'] =Reviews
```
